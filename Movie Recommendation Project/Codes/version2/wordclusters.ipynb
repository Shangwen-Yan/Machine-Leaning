```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from scipy.misc import imread
%matplotlib inline
```

```python
path = 'movielens-20m-dataset'
movie = pd.read_csv("./"+path+"/movie.csv")
movie = np.array(movie)
movies = np.loadtxt("./mvrcmdlist.txt",dtype=float).astype(int)
i = 1
```

```python
back_coloring = imread("cloud.jpg")
for clus in movies:
    m_clus = movie[clus[:80],2]
    wl_space_split = " ".join(m_clus)
    replaced = wl_space_split.replace("|"," ")
    result = ' '.join(replaced.split())
    my_wordcloud = WordCloud(background_color="white",mask=back_coloring).generate(result) 
    plt.figure(i)
    plt.imshow(my_wordcloud)
    plt.axis("off")
    plt.savefig("./wrdcnt/cluster%d.png"%(i))
    i +=1
```

```python
i = 1
for clus in movies:
    m_clus = movie[clus[:30],2]
    hist_data = "|".join(m_clus)
    hist_data = hist_data.split('|')
    df = pd.DataFrame({'genre':hist_data},index = range(len(hist_data)))
    group=df['genre'].groupby(df['genre'])
    genre = group.size().index
    freq = group.size().as_matrix()
    sum_f = sum(freq)
    freq = freq/sum_f
    
    plt.figure(figsize=(15,10))
    a = plt.bar(range(genre.size),freq,width=0.5)
    plt.xticks(range(genre.size),genre,rotation = 50)
    plt.grid()
    plt.savefig("./bar/bar%d.png"%(i))
    i +=1
```
